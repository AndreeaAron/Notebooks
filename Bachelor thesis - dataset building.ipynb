```python
import numpy as np
import pandas as pd
```

```python
fdi = pd.read_excel(r"C:\Users\Andreea\Documents\licenta\date\fdi.xlsx", sheet_name = 1)
fdi
```

```python
fdi.dropna(inplace = True)
fdi.dtypes
```

```python
infr = pd.read_csv(r"C:\Users\Andreea\Documents\licenta\date\infrastrcture.csv")
infr.dropna(inplace = True)
infr.dtypes
```

```python
infr['Partner iso'] = infr['Partner iso'].astype('object')
fdi['Partner iso'] = fdi['Partner iso'].astype('object')

df = pd.merge(fdi, infr, on='Partner iso')
df = df.drop('Country Name', axis = 1)
df
```

```python
ctl = pd.read_excel(r"C:\Users\Andreea\Documents\licenta\date\ctl + var control origin.xlsx")
df = pd.merge(df, ctl, on = 'Reporting ISO')
df
```

```python
ctl = pd.read_excel(r"C:\Users\Andreea\Documents\licenta\date\ctl + var control destination.xlsx")
df = pd.merge(df, ctl, on = 'Partner iso')
df
```

```python
df['tariff_dif'] = df.tariff_origin - df.tariff_destination
df = df.drop(['tariff_origin', 'tariff_destination'], axis = 1)
df
```

```python
df['ctl_dif'] = df.ctl_destination - df.ctl_origin
df
```

```python
gravity = pd.read_excel(r"C:\Users\Andreea\Documents\licenta\date\gravity.xlsx")
gravity = gravity.rename(columns = {'Country Origin':'Reporting ISO', 'Country Destination': 'Partner iso'})
gravity
```

```python
df = pd.merge(df, gravity, on = ['Reporting ISO','Partner iso'])
df
```

```python
df = df.dropna()
df
```

```python
pop = pd.read_excel(r"C:\Users\Andreea\Documents\licenta\date\population.xls")
pop = pop.rename(columns = {'Country Code': 'Partner iso'})
pop
```

```python
df1 = df.merge(pop, on = 'Partner iso')
df1['fdi/cap'] = df1['Average of fdi']/df1['Population']
df1
```

```python
gdp = pd.read_csv(r"C:\Users\Andreea\Documents\licenta\date\gdp\gdp.csv")
gdp = gdp.dropna()
gdp
```

```python
df1 = df1.merge(gdp, on = 'Partner iso')
df1['fdi/gdp'] = df1['Average of fdi']/df1['gdp']
df1 = df1.dropna()
```
